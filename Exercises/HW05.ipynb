# ***Connect to Firebase RealTime DB***

```python
!pip install firebase
from firebase import firebase
# Connect to Firebase Realtime Database
FBconn = firebase.FirebaseApplication('https://hw05db-default-rtdb.firebaseio.com/',None)
```

```python
def add_single_word(record_id):
  word_dict = FBconn.get('/Words/'+record_id, None)
  if word_dict is not None:
    FBconn.put('/Words/' + record_id, 'count', word_dict['count']+1)
  else:
    FBconn.put('/Words/' + record_id, 'count', 1)


def delete_record(record_id):
  """Delete a temperature record"""
  return FBconn.delete('/Words/', record_id)

def analyze_text(text):
  words = text.split()
  for word in words:
    add_single_word(word)

def update_word_count(record_id, count):
  word_dict = FBconn.get('/Words/'+record_id, None)
  if word_dict is not None:
    FBconn.put('/Words/' + record_id, 'count', count)
    print("Count updated successfully")
  else:
    print("Word not found")


def get_all_records():
  """Retrieve all temperature records"""
  return FBconn.get('/Words/', None)

def display_records():
  """Display all words records"""
  records = get_all_records() # Retrieve all records
  if records: # Check if there are records
    print("\nCurrent Records:")
    for record_id, data in records.items():
      print(f"{record_id} : {data['count']}")
  else:
    print("\nNo records found")
```

```python
while True:
  print("\nCommon Words Counter Menu")
  print("1. Add single word")
  print("2. Add text for analysis")
  print("3. Update word count")
  print("4. Delete word")
  print("5. View all words")
  print("6. Exit")

  choice = input("\nEnter your choice (1-6): ")


  if choice == '1':
    record_id = input("Enter word  to add: ")
    add_single_word(record_id)

  elif choice == '2':
    text = input("Enter text: ")
    analyze_text(text)

  elif choice == '3':
    record_id = input("Enter word: ")
    count = int(input("Enter new count: "))
    update_word_count(record_id, count)

  elif choice == '4':
   record_id = input("Enter word  to delete: ")
   delete_record(record_id)

  elif choice == '5':
   display_records()

  elif choice == '6':
    break

  else:
    print("\nInvalid choice. Please select 1, 2, 3, 4, or 5.")

print("\nBye")
```
